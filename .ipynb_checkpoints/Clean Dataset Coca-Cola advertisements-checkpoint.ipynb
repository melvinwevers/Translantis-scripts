```python
%matplotlib inline
import csv
import collections
import pprint
import sys
import pandas as pd
import matplotlib
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
import matplotlib.pyplot as plt
from pylab import figure, axes, pie, title, show
```

```python
csv.field_size_limit(sys.maxsize)
```

```python
main_file = "/Users/wevers/Data/output_filtered.csv" 
```

```python
df = pd.read_csv(main_file, header=0, sep='\t', names = ["date", "newspaper", "title", "content"])
#data = data.set_index(pd.DatetimeIndex(data['paper_dc_date']))
```

```python
df
```

```python
single = df.drop_duplicates(['title']) #remove duplicate articles based on article title
df2 = df.drop_duplicates(['title']) #remove duplicate articles based on article title
```

```python
single['newspaper'] = single['newspaper'].apply(lambda x: x.split(':')[0]) #strip cell after :
single['newspaper'] = single['newspaper'].apply(lambda x: x.split('/')[0]) #strip cell after /
single['newspaper'] = single['newspaper'].str.lower().str.replace('ĳ', 'ij') #fix error with ij
single = single.dropna()
```

```python
single.to_csv('clean.csv', index=False) #export to csv
```

```python
threshold = 150 # Anything that occurs less than this will be removed.
value_counts = single['newspaper'].value_counts()
to_remove = value_counts[value_counts <= threshold].index
single['newspaper'].replace(to_remove, inplace=True)
```

```python
plot = single['newspaper'].value_counts().plot(kind="bar") #plot newspaper_titles
```

```python
fig = plot.get_figure() #export plot to PNG file
fig.set_size_inches(18.5, 15.5)
fig.tight_layout()
fig.savefig('test2png.png', dpi=100)
```
