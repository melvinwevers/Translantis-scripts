```python
n_articles = 373
```

```python
main_file = "/Users/wevers/Data/amerika/1890-1899.txt" 
```

```python
df = pd.read_csv(main_file, header=0, sep=',', names = ["date","title", "content"])
#data = data.set_index(pd.DatetimeIndex(data['paper_dc_date']))
```

```python
df['content'] = df['content'].str.lower()
```

```python
word_regexs = [r'alcohol',
               r'apart*',
               r'aroma',
               r'best*',
               r'bijzonder*',
               r'[a-z]alorie*',
               r'chique',
               r'dagelijks',
               r'dorst',
               r'echte',
               r'extra',
               r'fijn',
               r'fles*',
               r'fris',
               r'gezellig*',
               r'gezinsfles*',
               r'glas',
               r'glazen',
               r'heerlijk',
               r'iedereen',
               r'kenner*',
               r'king?size',
               r'kingsize',
               r'kroonkurk*',
               r'kwaliteit',
               r'landen',
               r'lekker*',
               r'lescht',
               r'light',
               r'lijn',
               r'literfles*',
               r'modern*',
               r'mondain*',
               r'nieuw*',
               r'ontworpen',
               r'onvervangbare',
               r'opgewerkt*',
               r'opwekkend*',
               r'origine[a-z]*',
               r'overal',
               r'parel*',
               r'pauze',
               r'plastic',
               r'populair*',
               r'praktisch*',
               r'prijs',
               r'samenstelling',
               r'sierlijk*',
               r'smaak',
               r'sprankelen*',
               r'standaardfles*',
               r'stimuleer*',
               r'verfris*',
               r'verkrijgbaar',
               r'verkwik*',
               r'verukkelijk*',
               r'voordelig*',
               r'vruchten*',
               r'wereld',
               r'zuiver*',
               r'mil[a-z]oen*',
            ]
```

```python
rel_counts = pd.Series((df['content'].str.contains(r).sum() / n_articles * 100 for r in word_regexs), word_regexs, name='rel_count')
abs_counts = pd.Series((df['content'].str.contains(r).sum() for r in word_regexs), word_regexs, name='abs_count')
```

```python
abs_counts
```

```python
rel_counts = rel_counts.round(1)
```

```python
wordcount = pd.concat([rel_counts, abs_counts], axis=1)
```
