```python
%matplotlib inline
import pandas as pd
import matplotlib
import codecs
import datetime
```

```python
main_file = "taylorisme.csv" 
```

```python
df = pd.read_csv(main_file, header=1, skiprows=[1], sep="\t")
#data = data.set_index(pd.DatetimeIndex(data['paper_dc_date']))
```

```python
df #preview file
```

```python
df['text_content'] = df['text_content'].str.lower().str.replace('"', '""') #replace " with """" + lowercase text
```

```python
df['paper_dc_title'] = df['paper_dc_title'].apply(lambda x: x.split(':')[0]) #strip cell after :
df['paper_dc_title'] = df['paper_dc_title'].apply(lambda x: x.split('/')[0]) #strip cell after /
df['paper_dc_title'] = df['paper_dc_title'].str.lower().str.replace('ĳ', 'ij') #fix error with ij
```

```python
df['text_content'] = ('"') +  df['text_content'].map(str) + ('"') #add strings around text_content
```

```python
df['paper_dc_date'] = pd.to_datetime(df['paper_dc_date'], format="%Y-%m-%d") #turn datecolumn in date object
```

```python
df['paper_dc_date'] = df['paper_dc_date'].dt.strftime('%d.%m.%Y') #convert to desired format
```
